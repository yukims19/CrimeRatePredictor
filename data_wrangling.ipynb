```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

## Data Collection<a id='data_collection'></a>

```python
crime_raw = pd.read_excel('data/california_crime.xls', skiprows=4)
crime_raw.columns = crime_raw.columns.str.replace('\n', ' ')
crime_raw.head()
```

```python
city_raw = pd.read_csv('data/City_Boundaries.csv', index_col=0)
city_raw.head()
```

```python
health_raw = pd.read_csv('data/health_care.csv', index_col=0)
health_raw.head()
```

```python
school_raw = pd.read_csv('data/school.csv', index_col=2)
school_raw.head()
```

## Data Cleaning<a id='data_cleaning'></a>

```python
#crime_raw, city_raw, health_raw, school_raw
```

```python
#Check missing value
crime_raw.isna().sum()
na_row_idx = crime_raw[crime_raw['Violent crime'].isna()].index
crime_data = crime_raw.drop(na_row_idx)
print(crime_data.isna().sum())
```

```python
# Make city the index
crime_data['City'].duplicated().any()
crime_data.reset_index(inplace=True)
crime_data.set_index('City', inplace=True)
crime_data.index
crime_data.head()
```

```python
# Convert float to int
crime_data_int = crime_data.select_dtypes('float').astype('int')
crime_data_int.head()
```

```python
# change column names to snake case
crime_data_int.reset_index(inplace=True)
crime_data_int.columns = crime_data_int.columns.str.lower().str.replace(' ', '_').str.replace('-', '')
crime_df = crime_data_int
crime_df.head()
```

```python
# check missing data
print(city_raw.isna().sum())

# change column names to snake case
city_raw.columns = city_raw.columns.str.lower().str.replace(' ', '_')
city_no_index = city_raw.reset_index()
city_df = city_no_index.drop('OBJECTID', axis=1)
city_df.head()
```

```python
# check city_raw CITY values and crime_df City
print('city in city data: ', city_raw['city'].count(), 'city in crime : ', crime_df['city'].count())
city_city = city_raw['city']
crime_city = crime_df['city']
print('common count: ', city_city.isin(crime_city).sum(), crime_city.isin(city_city).sum())
cities_not_common = city_city[~city_city.isin(crime_city)]
cities_not_common_2 = crime_city[~crime_city.isin(city_city)]
print(cities_not_common, cities_not_common_2)
cities_common = city_city[city_city.isin(crime_city)]
```

```python
# merge by inner join on city
crime_cities = pd.merge(crime_df, city_df, on=['city'])
print(crime_cities.info())                            
```

```python
# change column names to snake case
health_raw.columns = health_raw.columns.str.lower().str.replace(' ', '_').str.replace('-', '')
health_no_index = health_raw.reset_index()
health_data = health_no_index.drop('_id', axis=1)

#Check missing value
print(health_data['city'].isna().sum())

#Check duplicate facilities
print(health_data.columns)
health_nodup = health_data.drop_duplicates(['facname', 'city', 'address', 'zip'])
health_nodup[['facname', 'city', 'address', 'zip']].duplicated().any()
missing_val_cols = health_nodup.columns[health_nodup.isna().sum() > 1000]
health_df = health_nodup.drop(missing_val_cols, axis=1).drop(['latitude','longitude','location'], axis=1)
```

```python
# change column names to snake case
school_raw.columns = school_raw.columns.str.lower().str.replace(' ', '_').str.replace('-', '')
school_no_index = school_raw.reset_index()
school_data = school_no_index.drop('OBJECTID', axis=1)

#Check missing value
print(school_data.isna().sum())

#Check duplicate facilities
school_nodup = school_data.drop_duplicates(['school', 'city', 'level_', 'street'])
school_nodup[['school', 'city', 'level_', 'street']].duplicated().any()
school_df = school_nodup.drop(['x','y','lat', 'long', 'notes', 'locations', 'cds_uniq', 'cdscode'], axis=1)
school_df.head()
```

## Data Definition<a id='data_definition'></a>

```python
# save data
school_df.to_csv('data/school_df.csv', sep=',')
health_df.to_csv('data/health_df.csv', sep=',')
crime_cities.to_csv('data/crime_cities_df.csv', sep=',')
```
